```python
# ==============================
# Task 3: Data Visualization
# Employee Dataset
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()
# ------------------------------
# 1. Load Dataset
# ------------------------------
df = pd.read_csv("employee_dummy_dataset.csv")

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# ------------------------------
# 2. Visualize Employee Count by Department
# ------------------------------
plt.figure(figsize=(8,5))
sns.countplot(x="Department", data=df, palette="Set2")
plt.title("Number of Employees per Department", fontsize=14)
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# ------------------------------
# 3. Gender Distribution
# ------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df, palette="pastel")
plt.title("Gender Distribution", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# ------------------------------
# 4. Performance Rating Distribution
# ------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Performance", data=df, palette="muted")
plt.title("Performance Rating Distribution", fontsize=14)
plt.xlabel("Performance Level")
plt.ylabel("Count")
plt.show()

# ------------------------------
# 5. Age Distribution
# ------------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=10, kde=True, color="skyblue")
plt.title("Age Distribution", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# ------------------------------
# 6. Salary Distribution
# ------------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["Salary"], bins=10, kde=True, color="salmon")
plt.title("Salary Distribution", fontsize=14)
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()

# ------------------------------
# 7. Experience vs Salary
# ------------------------------
plt.figure(figsize=(7,5))
sns.scatterplot(x="YearsExperience", y="Salary", hue="Performance", data=df, palette="deep", s=80)
plt.title("Years of Experience vs Salary", fontsize=14)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend(title="Performance")
plt.show()

# ------------------------------
# 8. Salary by Department
# ------------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x="Department", y="Salary", data=df, palette="Set3")
plt.title("Salary by Department", fontsize=14)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.xticks(rotation=30)
plt.show()

# ------------------------------
# 9. Salary vs Performance
# ------------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="Performance", y="Salary", data=df, palette="Set1")
plt.title("Salary Distribution by Performance Level", fontsize=14)
plt.xlabel("Performance")
plt.ylabel("Salary")
plt.show()

# ------------------------------
# 10. Correlation Heatmap (Numeric Columns)
# ------------------------------
plt.figure(figsize=(6,5))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```
